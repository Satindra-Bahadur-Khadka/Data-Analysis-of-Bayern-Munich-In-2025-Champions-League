```python
bayern_id = 50037
data = pd.read_csv('../Stats/players_data.csv')
bayern_data = data[data['id_team'] == bayern_id]
```

```python
bayern_data
```

```python
bayern_filtered = bayern_data[['id_player', 'player_name', 'nationality', 'field_position']]
```

```python
bayern_filtered = bayern_filtered.reset_index(drop=True)
bayern_filtered
```

```python
goals_data = pd.read_csv('../Stats/goals_data.csv')

goals_data_overall = goals_data[['id_player', 'goals']]
bayern_goals = pd.merge(bayern_filtered, goals_data_overall, on='id_player', how='left')

bayern_goals
```

```python
goals_per_player = goals_data.groupby('id_player', as_index=False)['goals'].sum()

bayern_goals = pd.merge(bayern_filtered, goals_per_player, on='id_player', how='left')

bayern_goals_scored = bayern_goals[bayern_goals['goals'] > 0]

bayern_goals_scored
```

```python
plt.Figure(figsize=(20,9))
plt.bar(bayern_goals_scored['player_name'], bayern_goals_scored['goals'], color='skyblue')
plt.title("Bayern Goal Scorers - 2025 Champions League")
plt.xlabel("Player's Name")
plt.ylabel("Goals Scored")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(bayern_goals_scored['goals'], labels=bayern_goals_scored['player_name'], autopct='%1.1f%%', startangle=140)
plt.title("Bayern Goal Distribution - 2025 Champions League")
plt.show()
```

```python
bayern_goals_scored = bayern_goals[bayern_goals['goals'] > 0]

goal_types = ['inside_area', 'outside_area']
bayern_goal_details = pd.merge(
    bayern_goals_scored,
    goals_data[['id_player'] + goal_types],
    on='id_player',
    how='left'
)

columns_order = ['player_name', 'goals'] + goal_types
bayern_goal_details = bayern_goal_details[columns_order]

bayern_goal_details
```

Here we filter out only the players who scored for proper analysis and to also analyse the area from which they scored. As we can see Harry Kane has the most amount of goals and all of them are from inside the Opponent's penalty area, since Harry is a center forward position player it is quite easy for him to achieve goals from inside the opponent's penalty box. Michael Olise the new Bayern Player has 2 goals to his name and both of them are inside the opponent's area as he is a brilliant dribbler. Other players have one goals to their names but the one who stands out among all is Leroy Sane the german player predominantly plays in the Left Wing and is the only one with a goal outside the area.

```python
attempts_data = pd.read_csv('../Stats/attempts_data.csv')
attempts_data.head()
```

Here we read the attempts_data.csv to check the formatting.

```python
bayern_attempts = pd.merge(bayern_filtered[['id_player', 'player_name']], attempts_data, on='id_player', how='left')

bayern_attempts_valid = bayern_attempts.dropna(subset=['total_attempts'])

bayern_attempts_valid
```

We filter out Bayern players from the attempts_data.csv based on the id_player column we merged out new bayern_filtered column and merged it with attempts_data by using id_players as foreign key to sort out the attempts data according to the players.

```python
bayern_attempts_valid = bayern_attempts_valid.sort_values(by='total_attempts', ascending=False)

plt.figure(figsize=(12,6))
plt.bar(
    bayern_attempts_valid['player_name'], 
    bayern_attempts_valid['total_attempts'], 
    color='orange'
)
plt.title("Bayern Players - Total Attempts (2025 Champions League)")
plt.xlabel("Player")
plt.ylabel("Total Attempts")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

In the bar graph above we can clearly see that Harry kane with the most number of attempts,followed by Michael Olise, Jamal Musiala, Serge Gnabry, Joshua Kimmich and so on. As we move further to the left we can see players who attempted to attack. Manuel Neuer and Sven Ulreich as they are the goal keepers they cannot attempt any attack, the same goes for the defenders Aleksander Pavlovic Minjae Kim. The one who stands out is Thomas Muller with no attempts to his name.

```python
attacking_data = pd.read_csv('../Stats/attacking_data.csv')

bayern_attacking = pd.merge( bayern_filtered[['id_player', 'player_name']], attacking_data, on='id_player', how='left')

bayern_attacking
```

From the code above we read the attacking_csv file to fetch the attacking data then we merge the data with bayern_filtered and attacking_data with reference to id_player in order to fetch the data corresponding to Bayern players. 

```python
bayern_attacking_valid = bayern_attacking.dropna(subset=['assists'])
bayern_attacking_valid
```

Here we clean the data ro only filter out players with any attacking data for proper analysis.

```python
attack_cols = ['assists', 'corners_taken', 'offsides', 'dribbles']

bayern_attacking_nonzero = bayern_attacking_valid[
    (bayern_attacking_valid[attack_cols] != 0).any(axis=1)
]

bayern_attacking_nonzero
```

Now we filter out players who had zero attacking in all aspects to only keep those who were involved in the competition.

```python
attack_cols = ['assists', 'corners_taken', 'offsides', 'dribbles']

bayern_attacking_nonzero.set_index('player_name')[attack_cols].plot(
    kind='bar', 
    stacked=True, 
    figsize=(12,6),
    colormap='tab20'
)

plt.title("Bayern Players - Attacking Stats Breakdown (2025 Champions League)")
plt.ylabel("Counts")
plt.xlabel("Player")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

From the graph above we can see that Joshua Kimmich and Michael Olise play and integral role when taking corners and we can conclude that their crosses and short corner passes are accurate. And we can also see that the prolific dribbler of Bayern is certainly Jamal Musialy as according to the data he has the most amount of dribbles to his name. We can also see that Thomas Muller has an assist to his name so we can also conclude that he was involved in the competition just not to an attacking level, unlike how the attempts_data suggested.

```python
defending_data = pd.read_csv('../Stats/defending_data.csv')


bayern_defending = pd.merge(bayern_filtered[['id_player', 'player_name']], defending_data, on='id_player', how='left')

bayern_defending
```

Here we import the defending data from defending_data.csv to analyse the defensive stats of Bayern.

```python
defense_cols = ['balls_recovered', 'tackles', 'tackles_won', 'tackles_lost', 'clearance_attempted']

bayern_defending_nonzero = bayern_defending.dropna(subset=defense_cols)
bayern_defending_nonzero = bayern_defending_nonzero[
    (bayern_defending_nonzero[defense_cols] != 0).any(axis=1)
]

bayern_defending_nonzero
```

Now we futher clean the data to remove the players who have no defending data and those who had 0 under every defensive aspect.

```python
bayern_defending_nonzero.set_index('player_name')[defense_cols].plot(
    kind='bar', 
    stacked=True, 
    figsize=(12,6),
    colormap='coolwarm'
)

plt.title("Bayern Players - Defensive Stats Breakdown (2025 Champions League)")
plt.ylabel("Counts")
plt.xlabel("Player")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

In the graphical illustration above we can see that Dayot Upamecano and Minjae Kim are the players who have the most activity in the defensive aspect of the team, with their ball recovery being remarkable. We can also see Joshua Kimmich who plays as a defensive centre mid-fielder being defensively proactive as well, along with Alphonso Davies who plays as a Left-back. The cause for concern based on this data is that the players who plays above the defensive line does not have that much to show in their defensive aspect. It is remarkable that the defenders have such good stats but the reason for that is because the forward playing players are doing so little to recover the ball back and leaving that responsibility to the defenders.

In conclusion, the data we have analysed suggests that the weakness that Bayern Munich have is in their activity of their players. Bundesliga and german players are known for their pressing and activity on the football pitch. The key parts where bayern should improve is when they lose the ball, instead of counting on the defenders to win the ball back and resume attacking they must focus on their offensive players to press the opposing players when the lose the posession. Bayern can also experiment with a high defensive line as Manuel Neuer performs the sweeper keeper role quite efficiently giving the players more chances to attack. Finally, when the players lose the possession of the ball the central midfield players can press the opponents to win the possession back while maintaining a center forward and attacking midfielder on offensive for a quick counter.
